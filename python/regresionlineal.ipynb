```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Suponiendo que ya tienes un DataFrame con tus datos
# Aquí solo usamos datos de ejemplo
data = {
    'positive_word_count': [10, 20, 30, 40, 50],
    'adjusted_rating': [3.5, 4.0, 4.2, 4.5, 4.7]
}

df = pd.DataFrame(data)

# Variables independientes y dependientes
X = df[['positive_word_count']]
y = df['adjusted_rating']

# Crear el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Coeficientes
coef = model.coef_[0]
intercept = model.intercept_

print(f'Coeficiente: {coef}')
print(f'Intercepto: {intercept}')

# Predicciones
df['predicted_rating'] = model.predict(X)

# Visualización
plt.scatter(df['positive_word_count'], df['adjusted_rating'], color='#71550a', label='Datos Reales')
plt.plot(df['positive_word_count'], df['predicted_rating'], color='#fbcd55', linewidth=2, label='Línea de Regresión')
plt.xlabel('Número de Palabras Positivas')
plt.ylabel('Calificación Promedio Ajustada')
plt.legend()
plt.show()
```
